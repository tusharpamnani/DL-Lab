```python
x_input = [0.1, 0.5, 0.2]
w_weight = [0.625,0.624,0.625]
threshold = 0.5
```

```python
def step(weighted_Sum, threshold) :
    if weighted_Sum >= threshold:
      return 1
    else:
      return 0
```

```python
def perecptron(x_input, w_weight):
  weighted_Sum = 0
  for x, w in zip(x_input, w_weight):
    weighted_Sum += x * w
    print(weighted_Sum)
  return step(weighted_Sum)
```

```python
output = perecptron(x_input, w_weight)
print(f'Output: {output}')
print(f'Optimized Weights: {w_weight}')
```

```python
w_weight=[0,0,0]
while((perecptron(x_input,w_weight))<=0):
   w_weight[0]+=0.00001
   w_weight[1]+=0.00001
   w_weight[2]+=0.00001
print(w_weight)
```

```python
def and_an(inputs):
  input_sum = 0
  threshold = len(inputs)
  for i in inputs:
    input_sum += i
  return step(input_sum, threshold)
```

```python
def and_func(inputs, weights):
  input_sum = 0
  threshold = 0
  bias = -1
  for i in range(len(inputs)):
    input_sum += inputs[i] * weights[i]
  return step(input_sum  + bias, threshold)
```

```python
inputs = [[0,0], [0,1], [1,0], [1,1]]
for i in inputs:
  weight = [1,1]
  op = and_func(i, weight)
  print(i, op)
```
